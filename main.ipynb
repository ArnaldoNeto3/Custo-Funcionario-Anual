```python
import pandas as pd

df = pd.read_csv('base de dados.csv')
dados = df.to_dict(orient='records')

horas_mensais = {}
horas_anuais = {}
salarios = {}
custo_anual = {}

for dado in dados:
    nome = dado['Nome']
    mes = dado['Mes']
    horas = dado['HH - Trabalhado']
    salario = dado['Salario']

    # Guardar salário do funcionário (pega apenas o primeiro encontrado)
    if nome not in salarios:
        salarios[nome] = salario
   
    if nome not in horas_mensais:
        horas_mensais[nome] = {}

    if mes not in horas_mensais[nome]:
        horas_mensais[nome][mes] = 0

    if nome not in horas_anuais:
        horas_anuais[nome] = 0

    # Somatórios
    horas_mensais[nome][mes] += horas
    horas_anuais[nome] += horas

# Calculando custo anual de cada funcionário

for nome in horas_anuais:
    valor_hora = salarios[nome] /220
    custo_anual[nome] = valor_hora * horas_anuais[nome]

for nome in custo_anual:
    print(f"{nome} R${custo_anual[nome]:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
```
